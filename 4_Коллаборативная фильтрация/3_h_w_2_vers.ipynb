```python
from surprise import Dataset
from surprise.model_selection import cross_validate
from surprise import KNNBasic, KNNWithMeans, KNNWithZScore, KNNBaseline
from surprise import SVD, SVDpp, NMF, SlopeOne, CoClustering
from surprise import Reader
import pandas as pd

movies = pd.read_csv("D:/Documents/python_projects/Netology/venv/Рекомендательные системы/1_Введение и классификация рекомендательных систем/movies.csv")
ratings = pd.read_csv("D:/Documents/python_projects/Netology/venv/Рекомендательные системы/1_Введение и классификация рекомендательных систем/ratings.csv")
```

```python
movies_with_ratings = movies.merge(ratings, on='movieId').reset_index(drop=True)
movies_with_ratings.dropna(inplace=True)
movies_with_ratings.head()
```

```python
dataset = pd.DataFrame({
    'uid': movies_with_ratings.userId,
    'iid': movies_with_ratings.title,
    'rating': movies_with_ratings.rating
})
```

```python
dataset.head()
```

```python
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(dataset, reader)
```

```python
data
```

```python
models = {
    "KNNBasic": KNNBasic(),
    "KNNWithMeans": KNNWithMeans(),
    "KNNWithZScore": KNNWithZScore(),
    "KNNBaseline": KNNBaseline(),
    "SVD": SVD(),
    "SVDpp": SVDpp(),
    "NMF": NMF(),
    "SlopeOne": SlopeOne(),
    "CoClustering": CoClustering()
}
list_models = []

for name, model in models.items():
    print(f"\n=== Evaluating {name} ===")
    cv_results = cross_validate(
        model, 
        data, 
        measures=["RMSE"], 
        cv=5,
        verbose=True,
        n_jobs=-1
    )
    
    avg_rmse = sum(cv_results["test_rmse"]) / 5
    if avg_rmse <= 0.87:
        list_models.append(name)
        print(f"\nAverage RMSE: {avg_rmse:.4f}")
    else:
        print(f"Модель {name} не подходит")
print (f"Список моделей с RMSE <= 0,87 {list_models}")
```
